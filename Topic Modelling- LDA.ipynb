Reading in the data

```python
import pandas as pd
headlines = pd.read_csv("abcnews-date-text.csv")
headlines = headlines.drop(columns = "publish_date")
headlines = headlines.head(50)
#print(headlines)
```

```python
Cleaning the data
```

```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string 
exclude = set(string.punctuation)
stop = set(stopwords.words("english"))
lemma = WordNetLemmatizer()
tokens_list = []
for i in range(len(headlines)):
    stop_free = " ".join([i for i in headlines.headline_text[i].lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    k = (normalized.split())
    tokens_list.append(k)
    
#print(tokens_list)
```

Building the dictionary and document-term matrix

```python
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
import gensim
from gensim import corpora
dictionary = corpora.Dictionary(tokens_list)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in tokens_list]
print(doc_term_matrix)
```

Building the model

```python
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(doc_term_matrix, num_topics = 20,id2word = dictionary, passes = 50)
```

Getting the topics

```python
import pandas as pd
def get_lda_topics(model, num_topics):
    word_dict = {};
    for i in range(num_topics):
        words = model.show_topic(i, topn = 15);
        word_dict['Topic # ' + '{:02d}'.format(i+1)] = [i[0] for i in words];
    return pd.DataFrame(word_dict);
get_lda_topics(ldamodel,10)
```
